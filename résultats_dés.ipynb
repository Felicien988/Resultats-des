```python
import pandas as pd
def moyenne_dés(dés_lancés, dés_gardés) :
  jet = []
  resultat = []
  i=0
  j=0
  while i<dés_lancés :
    jet.append(1)
    i=i+1
  while j<6**dés_lancés :
    k=0
    while k<dés_lancés :
      jet[k] = int(j/6**k+10**-12)%6+1
      k=k+1
    jet.sort()
    resultat.append(sum(jet[-dés_gardés:]))
    j=j+1
  moyenne = sum(resultat)/len(resultat)
  result = pd.DataFrame({"résultat" : resultat})
  resultats = result.groupby("résultat")["résultat"].count().reset_index(name='Compte')
  resultats['proba(%)'] = resultats['Compte'] / resultats['Compte'].sum() * 100
  resultats['proba_cumulée(%)'] = 100-resultats['proba(%)'].cumsum()+resultats['proba(%)']
  resultats = resultats.set_index('résultat')
  proba_max = max(resultats['proba(%)'])
  max_index = resultats.idxmax()['Compte']
  return {"moyenne" : moyenne, "proba_max" : proba_max, "max_index" : max_index, "resultats" : resultats}

```

```python
moyenne_dés(5,2)
```

```python
df = pd.read_excel("/content/dés.xlsx")
df
```

```python
i=1
while i <=9 :
  j=1
  while j <= min(i,6) :
    dico = moyenne_dés(i,j)
    df.iloc[3*j-3,i+1] = dico["moyenne"]
    df.iloc[3*j-2,i+1] = dico["max_index"]
    df.iloc[3*j-1,i+1] = dico["proba_max"]
    j=j+1
  i=i+1
```
